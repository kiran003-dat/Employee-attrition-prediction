```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('/content/WA_Fn-UseC_-HR-Employee-Attrition.csv')
dataset.head()
```

```python
dataset.drop(columns=['EmployeeCount', 'EmployeeNumber', 'StandardHours', 'Over18'], inplace=True)
dataset.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
categorical_cols = dataset.select_dtypes(include=['object']).columns
numerical_cols = dataset.select_dtypes(exclude=['object']).columns
encoder = OneHotEncoder(sparse_output=False, drop='first')
# Fit and transform the categorical columns
encoded_data = encoder.fit_transform(dataset[categorical_cols])

# Convert the encoded data into a DataFrame with appropriate column names
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_cols))

# Combine the numerical data and the encoded categorical data
final_dataset = pd.concat([dataset[numerical_cols].reset_index(drop=True), encoded_df], axis=1)
```

```python
final_dataset.shape
```

```python
final_dataset = final_dataset.rename(columns={'Attrition_Yes': 'Attrition'})
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

x = final_dataset.drop(columns='Attrition')  # Independent variables
y = final_dataset['Attrition']  # Target variable

# Apply SelectKBest
bestfeatures = SelectKBest(score_func=f_classif, k=10)  # k is the number of top features to select
fit = bestfeatures.fit(x, y)

# Get scores of the features
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x.columns)

# Concatenate the scores and feature names into a single DataFrame
feature_scores = pd.concat([dfcolumns, dfscores], axis=1)
feature_scores.columns = ['Feature', 'Score']
```

```python
feature_scores_df = pd.DataFrame({
    'Feature': x.columns,
    'Score': feature_scores['Score']
})

# Sort by score for better visualization (optional)
feature_scores_df = feature_scores_df.sort_values(by='Score', ascending=False)

# Step 4: Plot the feature scores using a vertical bar plot
plt.figure(figsize=(10, 6))
plt.bar(feature_scores_df['Feature'], feature_scores_df['Score'], color='skyblue')
plt.ylabel('Feature Score')
plt.xlabel('Features')
plt.title('Feature Scores from Univariate Selection')
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels for better readability
plt.ylim(0, feature_scores_df['Score'].max() * 1.1)  # Optional: Add some space above the highest bar
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
```

```python
top_features = feature_scores.nlargest(35, 'Score')  # Select top 3 features

# Get the feature names
selected_feature_names = top_features['Feature'].values

# Filter the dataset to keep only the top features
x_selected = x[selected_feature_names]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_selected , y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
# Create a Logistic Regression model
model_1 = LogisticRegression(random_state=0)

# Fit the model to the training data
model_1.fit(x_train, y_train)

# Make predictions on the test data
y_pred = model_1.predict(x_test)

conf_matrix = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

results = pd.DataFrame([['Logistic Regression', acc, prec, rec, f1]],
               columns = ['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score'])
results
```

```python
from sklearn.model_selection import KFold, cross_val_score
kfold = KFold(n_splits=5, shuffle=True, random_state=1)

# Step 3: Perform Cross Validation
cv_results = cross_val_score(model_1, x_train, y_train, cv=kfold, scoring='accuracy')
print("Cross-Validation Accuracies:", cv_results)
print("Mean Accuracy:", cv_results.mean())
```
